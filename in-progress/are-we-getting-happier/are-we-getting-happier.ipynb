```julia
using DataFrames
import Datetime: date

@vectorize_1arg String date

tmpData = readtable("dailyFull.csv", header=false, skipstart=1)
data = DataFrame(date=date(tmpData[:, :x1]), value=tmpData[:, :x2])
head(data)
```

```julia
using Gadfly

set_default_plot_size(24cm, 16cm)
colorGradient = Scale.lab_gradient(color("crimson"), color("blue"), color("greenyellow"))
plot(
    data, 
    x = "date", 
    y = "value", 
    color = "value", 
    Guide.xlabel("Date"), 
    Guide.ylabel("Average Happiness"), 
    Guide.title("Average Happiness on Twitter"),
    Scale.ContinuousColorScale(colorGradient)
)
```

```julia
using JuMP
using GLPKMathProgInterface

m = Model(solver=GLPKSolverLP())
@defVar(m, x0)
@defVar(m, x1)
@defVar(m, x2)
@defVar(m, x3)
@defVar(m, x4)
@defVar(m, x5)

errors = Array(Variable, nrow(data))

function predict(a1)
    a2 = cos(2pi*a1/365.25)
    a3 = sin(2pi*a1/365.25)
    a4 = cos(2pi*a1/(10.7 * 365.25))
    a5 = sin(2pi*a1/(10.7 * 365.25))
    x0 + a1*x1 + a2*x2 + a3*x3 + a4*x4 + a5*x5
end

for a1 = 1:nrow(data)
    @defVar(m, deviation)
    prediction = predict(a1)
    
    @addConstraint(m, deviation >= prediction - data[a1,:value])
    @addConstraint(m, deviation >= data[a1,:value] - prediction)
    
    errors[a1] = deviation
end

@setObjective(m, Min, sum(errors))

status = solve(m)

x0 = getValue(x0)
x1 = getValue(x1)
x2 = getValue(x2)
x3 = getValue(x3)
x4 = getValue(x4)
x5 = getValue(x5)

println("Least-Absolute-Deviation = ", getObjectiveValue(m))

println("x0 = ", x0)
println("x1 = ", x1)
println("x2 = ", x2)
println("x3 = ", x3)
println("x4 = ", x4)
println("x5 = ", x5)
```
